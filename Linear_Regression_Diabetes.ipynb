```python
import numpy as np
db = np.loadtxt("0000000000002329_training_diabetes_x_y_train.csv", delimiter=",")
db.shape
```

```python
X=db[:,0:-1]
Y=db[:,:-1]
print(X)
print(Y)
```

```python
from sklearn.linear_model import LinearRegression
alg=LinearRegression()
alg.fit(X,Y)
```

```python
alg.coef_
```

```python
alg.intercept_
```

```python
db_test=np.loadtxt("0000000000002329_test_diabetes_x_test.csv",delimiter=',')
print(db_test.shape)
db_predict=alg.predict(db_test)
print(db_predict.shape, db_test.shape)
np.savetxt('results.csv',db_predict,fmt='%1.5f')
```

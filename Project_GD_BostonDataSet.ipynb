```python
import numpy as np
import pandas as pd
```

```python
boston = np.loadtxt("0000000000002417_training_boston_x_y_train.csv", delimiter=",")
X_train = boston[:, 0:-1]
Y_train = boston[:, -1]
print(X_train.shape)
print(Y_train.shape)
X_test = np.loadtxt("0000000000002417_test_boston_x_test.csv", delimiter=",")
print(X_test.shape)
```

```python
def step_gradient(x, y, learning_rate, m):
    M = x.shape[0]
    m_slope = (-2/M)*(((y - (x*m).sum(axis=1)).reshape(-1, 1))*x).sum(axis=0)
    m = m - learning_rate * m_slope
    return m
```

```python
def gd(x, y, learning_rate, num_iterations):
    m = np.zeros(x.shape[1])
    for i in range(num_iterations):
        #calculate m for each iteration
        m = step_gradient(x, y, learning_rate, m)
        print(i, "Cost :", cost(x, y, m))
    return m  
```

```python
def cost(x, y, m):
    return ((y - (x*m).sum(axis=1)) ** 2).mean()
```

```python
def run(x, y):
    df = pd.DataFrame(x)
    #add another column having all values as 1
    df[x.shape[1]] = 1
    #pass this to the gd function
    x = df.values
    learning_rate = 0.15
    num_iterations = 1000
    m = gd(x, y, learning_rate, num_iterations)
    return m
m = run(X_train, Y_train)  
```

```python
def result(x,m):
    return (x*m).sum(axis=1)
```

```python
df = pd.DataFrame(X_test)
del df[13]
del df[14]
print(df)
df[X_test.shape[1]] = 1
X_test = df.values
y_pred = result(X_test, m)
print(y_pred)
```

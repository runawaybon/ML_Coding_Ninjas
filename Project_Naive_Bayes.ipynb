```python
x = [] 
y = []
for folder in os.listdir(r"C:\Users\daddy\Desktop\20_newsgroups"):
    for doc in os.listdir(r"C:\Users\daddy\Desktop\20_newsgroups\\" + folder):
        with open(r"C:\Users\daddy\Desktop\20_newsgroups\\"+folder+'\\'+doc, "r") as f:
            x.append((doc,f.read()))
            y.append(folder)
```

```python
len(x[1249][0])
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(x,y, random_state = 0)
```

```python
dic={}
for i in range(len(x_train)):
    ds=x_train[i][1].lower()
    words=re.split(r'\W+',ds)
    for word in words:
        if not(word.isalpha()) or word in stop_words or len(word)<=2:
            continue
        if word in dic:
            dic[word]+=1
        else:
            dic[word]=1
```

```python
words_sorted = sorted(dic.items(), key=operator.itemgetter(1),reverse=True)
words_sorted
```

```python
features=words_sorted
answer1=[]
answer2=[]
for i in range(len(features)):
    answer1.append(i)
    answer2.append(features[i][1])
plt.plot(answer1,answer2)
plt.axis([0,8000,1,5000])
plt.grid()
plt.show()
```

```python
import re
import nltk
from nltk.corpus import stopwords
```

```python
def tweet_to_words(orig_tweet):
    letters_only = re.sub("[^a-zA-Z]", " ",orig_tweet) 
    words = letters_only.lower().split()                             
    stops = set(stopwords.words("english"))                  
    words_final = [w for w in words if not w in stops] 
    return( " ".join( words_final ))
```
